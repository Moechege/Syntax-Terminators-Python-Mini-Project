```python
#since the three datasets contain the same number of rows and columns,it is easier to concatenate and clean then analyze the data
#concatenate data 
telcom_df=pd.concat([df1,df2,df3],axis=0)
telcom_df
```

```python
#Clean data by , dropping unnecessary columns, checking for outliers, checking for duplicate values, handling missing data and checking for unique values
#drop unnecessary columns 
telcom_df.drop(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE'],axis=1)
```

```python
#check for outliers.Since the value column is the only quantitative column, we will check for outliers here 
telcom_df.VALUE.mean()
```

```python
telcom_df.VALUE.std()
```

```python
upper_limit=telcom_df.VALUE.mean() + 3*telcom_df.VALUE.std()
upper_limit
```

```python
lower_limit=telcom_df.VALUE.mean()-3*telcom_df.VALUE.std()
lower_limit
```

```python
telcom_df[(telcom_df.VALUE>upper_limit) | (telcom_df.VALUE<lower_limit)]
```

```python
#remove outliers 
telcom_df_no_outlier=telcom_df[(telcom_df.VALUE<upper_limit) & (telcom_df.VALUE>lower_limit)]
telcom_df_no_outlier
```

```python
#check number of outliers removed 
telcom_df.shape[0]-telcom_df_no_outlier.shape[0]
```

```python
#geo data set 
df_cells_geo=pd.read_csv('C:\\Users\\mac\Downloads\cells_geo - cells_geo.csv')
#split the dataset
df_cells_geo=df_cells_geo[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].str.split(pat=';',n=11, expand=True)
#drop the column 0
df_cells_geo.drop([0],axis=1,inplace=True)
#rename the columns 
df_cells_geo.columns=['VILLES','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','REGION','AREA','CELL_ID','SITE_CODE']
#replace empty whitespace with NaN
df_cells_geo.replace(r'^\s*$', np.nan, regex=True, inplace=True)
df_cells_geo
```
